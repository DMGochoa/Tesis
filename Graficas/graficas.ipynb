## Tipos de distribuciones

A continuación se va a realizar las gráficas que tendrian distintas distribuciones ejemplo para el la subsección de Datos Atípicos del marco teórico.

```python
# Parámetros para distribución normal
mu, sigma = 0, 0.2
datos0 = np.random.normal(mu, sigma, 1000)
```

```python
# Histograma 
sns.histplot(datos0, bins=20, kde=True)
#cuenta, cajas, _ = plt.hist(datos, 20, rwidth=0.92)
plt.title('Normal')
```

```python
# Distribución Uniforme
datos1 = np.random.uniform(0, 10, size=1000)
```

```python
sns.histplot(datos1, bins=20, kde=True)
plt.title("Uniform")
```

```python
# Graficando Geométrica
p =  0.3 # parametro de forma 
geometrica = np.random.geometric(p, 1000)
```

```python
sns.histplot(geometrica, bins=20, kde=True)
plt.title("Positively skewed")
```

```python
sns.histplot(abs(geometrica - geometrica.max() + 1), bins=20, kde=True)
plt.title("Negatively skewed")
```

```python
valores = np.random.standard_cauchy(1000)
valores = valores[(valores>-25) & (valores<30)]
sns.histplot(valores, bins=20, kde=True)
plt.title("Heavy tailed")
```

```python
x0 = np.random.standard_cauchy(1000)
x0 = x0[(x0>-25) & (x0<30)]
x1 = np.random.uniform(-25, 30, x0.__len__())
salida = np.sort(x1) - np.sort(x0)
sns.histplot(salida, bins=20, kde=True)
plt.title("Thin tail")
```

```python
fig, axes = plt.subplots(3, 2, figsize=(14, 16))

sns.histplot(datos0, bins=20, kde=True, ax=axes[0, 0])
axes[0, 0].set_title("Normal",
                    fontsize=15)
sns.histplot(datos1, bins=20, kde=True, ax=axes[0, 1])
axes[0, 1].set_title("Uniforme",
                    fontsize=15)
sns.histplot(geometrica, bins=20, kde=True, ax=axes[1, 0])
axes[1, 0].set_title("Asimétrica positiva",
                    fontsize=15)
sns.histplot(abs(geometrica - geometrica.max() + 1), bins=20, kde=True, ax=axes[1, 1])
axes[1, 1].set_title("Asimétrica negativa",
                    fontsize=15)
sns.histplot(valores, bins=20, kde=True, ax=axes[2, 0])
axes[2, 0].set_title("Cola pesada",
                    fontsize=15)
sns.histplot(salida, bins=20, kde=True, ax=axes[2, 1])
axes[2, 1].set_title("Cola liviana",
                    fontsize=15)

plt.suptitle("Probability Distributions",
            position=(0.5, 0.92),
            fontsize=18,
            fontweight="bold")

fig.savefig("tipodistri.svg")
```
